First, we read the data from the Excel spreadsheet given, into a pandas DataFrame named "price_df".

```python
import pandas as pd
import numpy as np

price_df = pd.read_excel('GWP_PTAP_Data.xlsx', sheet_name="10 SPDRs and S&P 500", index_col=0, header=1)

print("DataFrame displaying monthly prices:")
price_df
```

## a) Compute the price return for each SPDR and the S&P 500 for each month of 2017.

We calculate log returns for each SPDR and S&P, dividing the current month's price by the previous month, and then taking the natural log of that result. The results are collected in the "return_df" pandas DataFrame.

```python
return_df = np.log(price_df/price_df.shift(1)).dropna()
print("DataFrame displaying log monthly returns:")
return_df
```

## b) Compute the active return (relative to the S&P 500) for each SPDR for each month of 2017.

The active return is defined as the difference between a portfolio's return and the benchmark's return. In this case, the benchmark is given as the S&P 500. We create and store the active returns calculated in "active_return_df".

```python
active_return_df = pd.DataFrame(index=return_df.index)
```

```python
columns = list(return_df)
del columns[-1]

for i in columns:
    active_return_df[i+"_act_ret"] = return_df[i] - return_df['S&P 500']
```

```python
print("DataFrame displaying monthly active returns:")
active_return_df
```

## c) Calculate the monthly tracking error for each SPDR in 2017.

Tracking error is defined as the standard deviations of active returns. Thus, we calculate the standard deviations of active returns from "active_return_df", and then sort the tracking error in ascending order, so that we can easily determine the SPDR which has the lowest tracking error.

```python
print("Pandas series showing monthly tracking error:")
active_return_df.std(axis=0).sort_values()
```
